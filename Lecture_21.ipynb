# Different ways to import a library

The main concept of this particular lecture which is different ways to import a library.
Now we are introduced one more library function is called calender.

```python
import calendar
print(calendar.month(2026,2))
# first print statement allows us calender for a month.
# second print statement allows us calender for a year.
print(calendar.calendar(2026))
```

Earlier we were writting import calendar, then we say import calendar brings the entire library library of calendar into this python program and anything can be accessed only through this particular word calendar.
Due to which we always have to write calendar dot calendar or calendar dot month as we saw earlier
But in this second case from calendar import *, which means we are telling the computer to bring everything inside the calendar library into this particular program.* is used to all the feature of library.

```python
from calendar import*
print(month(2026,3))
print(calendar(2026))
```

The last type of import statement. In which we can import a specific library or a specific feature of the library an store it in  a variable.

```python
import calendar as c
print(c.month(2026,3))
```

```python
from calendar import month as m
# particular calendar library has a feature month and it stored in a variable m.
print(m(2026,3))
```
