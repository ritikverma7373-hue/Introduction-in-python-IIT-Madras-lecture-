# Escape characters and types of quotes

```python
print('It's a beautiful day')
```

Let us execute. It says invalid syntax, because over here, It s has this apostrophe and computer does not differentiate between this apostrophe and the single quote, which we usually use to start the string and stop the string.
How to solve this problem.Our easiest solution is to make it, it is and it will work. But I do not want to do that,Iwant to keep it like and this problem ar this particular error can be fixed using escape character.


Escape character is a mechanism which allows us to add those specific symbols in a string in print,which usually we can not do, as you see over here, with this single quote. Escape character is written using a backslash followed by the character, which you want to insert in the particular string.

```python
print('It\'s a beautiful day')
print("we are from \"IIT\" Madras")
```

In a computer we will use tab to gap in a sentence, but in a python language we will use(\t) for gap if you can see more  gap between sentences. if we add one more t over here,this escape character is called tab.

```python
print('My name is Ritik.\t\t I am from orai')
```

One more Escape character is (\n) backslash n, n stands for new line.

```python
print('My name is Ritik.\nI am from orai.')
```
