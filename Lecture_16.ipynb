```python
a_1=7
print(a_1)
AZ=90
print(AZ)
aZS_1=0
print(aZS_1)
```

# Shorthand Operators

shorthand operators are kind of airthmatic operators but htet have unique feature

```python
count=0
count=count+1
count=count+1
count=count+1
print(count)
```

But writing this name of the variable count again and again, is bit tedious and unnecessary,this is why Python has to this particular problem and the solution is called shorthand writting.

```python
count=0
count+=1
count+=1
count+=1
print(count)
```

shorthand operator use it only with addition operator or is it possible to use with other airthmatic operator as well?
The answer is, yes it is possible to use such a shorhand operator with all airthmatic operators.

```python
count=10
count-=1
count=count-1
print(count)
count=10
count*=2
count=count*2
print(count)
```
