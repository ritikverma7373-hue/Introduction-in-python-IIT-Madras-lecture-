# Introduction to "import library"

5 factorial is 120; 5 into 4 into 2 into 1, so on and so forth. so you have this facility on the calculator but how do we get such a facility on a computer, let us say in python.
This is an example to explain what is called library functions in Python.

```python
import math
print(math.factorial(5))
print(math.sqrt(16))
print(math.pow(10,3))
```

```python
# Let us to simulate the sum of two dies(six faces).

import random
dice1=(random.randrange(1,7))
# In just previous line says range 1 to 7, the range start with first but goes to last-1 element.
dice2=random.randrange(1,7)
total=dice1+dice2
print("your pair of dice is:",total)
```
