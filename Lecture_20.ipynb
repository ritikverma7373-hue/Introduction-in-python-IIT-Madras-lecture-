# Tutorial on if, else and else-if(elif) conditions

Find whether the given number is even or odd

```python
num=int(input("Enter a number"))
if(num%2==0):
  print('even')
else:
  print('odd')
```

Find whether the given number ends with 0 and 5 or any other number

```python
num=int((input('Enter a number: ')))
if (num%5==0):
  if(num%10==0):
     print('0')
  else:
     print('5')
else:
   print('other')
```

Find the garde of student based on the given marks from 0 to 10

Who are counted for A grade, does not appear again inB grade. This particular thing cab be acheived using a different kind of conditional statement called else-if,(where the second block executed only if first block does not satisfy the given condition.

```python
marks=int(input('Enter a number: '))
if (0<=marks<=100):
  if (marks>=90):
    print('A')
  elif(marks>=80):
    print('B')
  elif(marks>=70):
    print('C')
  elif(marks>=60):
    print('D')
  else:
    print('E')
else:
  print('Invalid input')
```
